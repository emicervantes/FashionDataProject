### Text Analysis

* Use NLP built-in functions to perform text analysis on `review_summary`

```python
def processText(text):
    text = text.str.replace("(<br/>)", "")
    text = text.str.replace('(<a).*(>).*(</a>)', '')
    text = text.str.replace('(&amp)', '')
    text = text.str.replace('(&gt)', '')
    text = text.str.replace('(&lt)', '')
    text = text.str.replace('(\xa0)', ' ')
    return text
```

```python
df2['review_text'] = processText(df2['review_text'])

df2['polarity'] = df2['review_text'].map(lambda txt: TextBlob(txt).sentiment.polarity)
df2['reviewLen'] = df2['review_text'].astype(str).apply(len)
df2['wordCount'] = df2['review_text'].apply(lambda x: len(str(x).split()))
```

```python
# List out reviews based on polarity
def listPolarity(df, pol, col_name, n):
    col = df.loc[df.polarity == pol, [col_name]].sample(n).values
    for c in col:
        print(c[0])
        
print("Reviews with the highest positive sentiment polarity: ")
print("-----------------------------------------------------")
listPolarity(df2, 1, 'review_text', 5)
print("")
print("Reviews with most neutral sentiment polarity: ")
print("-----------------------------------------------")
listPolarity(df2, 0, 'review_text', 5)
print("")
print("Reviews with the highest negative sentiment polarity: ")
print("------------------------------------------------------")
listPolarity(df2, -1, 'review_text', 5)
```

### Data Visualization

```python
# Histogram of polarity scores

score_pol = np.array(df2['polarity'])

fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
plt.hist(x = score_pol, color = 'pink', edgecolor = 'black', bins = 15)
plt.title("Polarity Score Distribution of Review Content", weight = "bold", fontsize = 15)
plt.xlabel("Polarity Score", weight = "bold", fontsize = 12)
plt.ylabel("Count", weight = "bold", fontsize = 12)
plt.show()
```

```python
# Dataset for each clothing category
dfDress = df2.loc[df['category'] == "dresses"]
dfTops = df2.loc[df['category'] == "tops"]
dfBottoms = df2.loc[df['category'] == "bottoms"]
dfOuter = df2.loc[df['category'] == "outerwear"]
```

```python
# Review polarity distribution for cloth category
# Histogram of polarity scores
polD = np.array(dfDress['polarity'])
polB = np.array(dfTops['polarity'])
polT = np.array(dfBottoms['polarity'])
polO = np.array(dfOuter['polarity'])
# dress -
plt.figure(figsize=(100, 70))
fig, (ax1, ax2, ax3, ax4)  = plt.subplots(1, 4, sharey='row', figsize=(20, 7))
ax1.hist(x = polD, color = 'pink', edgecolor = 'black', bins = 15)
ax1.set(xlabel = "Polarity Score", ylabel = "Count", title = "Dress")
# bottoms - 
ax2.hist(x = polB, color = 'lightblue', edgecolor = 'black', bins = 15)
ax2.set(xlabel = "Polarity Score", ylabel = "Count", title = "Bottoms")
# tops - 
ax3.hist(x = polT, color = 'lightgreen', edgecolor = 'black', bins = 15)
ax3.set(xlabel = "Polarity Score", ylabel = "Count", title = "Tops")
# outer - 
ax4.hist(x = polO, color = 'tan', edgecolor = 'black', bins = 15)
ax4.set(xlabel = "Polarity Score", ylabel = "Count", title = "Outerwear")

plt.legend()
fig.suptitle("Polarity Score Distribution of Review Content", weight = "bold", fontsize = 15)
plt.show()
```
